# Exploratory Data Analysis

### - Finding the most popular TED talks

```python
most_popular_ted_talks = ted_talks.sort_values("views", ascending=False).head(20)
most_popular_ted_talks.head(10)
```

## The most popular TED talks

```python
#THE MOST POPULAR TED TALKS 
plt.figure(figsize=(5,15))
sns.barplot(x='views', y='title', data=most_popular_ted_talks)
plt.show()
```

## Finding the most popular TED talks Speaker (in terms of number of talks)

## THE MOST POPULAR TED TALKS SPEAKER

```python
author = ted_talks['author'].value_counts().sort_values(ascending = False).index[:10]
author_df = pd.DataFrame({'author': author})
plt.figure(figsize=(15,10))
sns.countplot(x = 'author', data = author_df, order = author)
plt.title('Popularity of top 10 TED Talk Speakers based on their number of speeches')
plt.xlabel('Authors')
plt.ylabel('No. of TED Talks')
plt.show()
```

### Finding TED talks of your favorite Author

```python
my_favorite = ted_talks[ted_talks['author'] == 'Bill Gates']
my_favorite
plt.figure(figsize=(10, 15))
sns.barplot(x='title', y='views', data=my_favorite)
plt.title('TED Talks by Steve Jobs')
plt.xlabel('TED Talks')
plt.ylabel('No. of Views')
plt.xticks(rotation = 45)
plt.show()
```
